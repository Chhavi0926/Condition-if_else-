```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
market_share_percentages = [30, 25, 20, 15, 10]

# Create a Pandas DataFrame
market_data = pd.DataFrame({
    'Product': products,
    'Share': market_share_percentages
})

# Display the created DataFrame
print("Market Share Data:")
display(market_data)
```

```python
# Set figure size for better visualization
plt.figure(figsize=(8, 8))

# Create the pie chart
# 'labels' assigns product names to slices
# 'autopct' formats the percentage displayed on each slice
# 'startangle' rotates the start of the first slice
plt.pie(market_data['Share'], labels=market_data['Product'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

# Add a title to the pie chart
plt.title('Market Share Distribution by Product')

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')


# Display the chart
plt.show()
```
